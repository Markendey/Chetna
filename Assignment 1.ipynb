```python
#check whether the number is prime or not
def prime(n):
    for i in range(2, n):
         if n%i==0 or n<=1:
                return False
    return True
```

```python
#recursive function
def fact(n):
    if n==0 or n==1:
        return 1
    return n*fact(n-1)
```

```python
#choosing option whether to sum or product the given input
def sum(n):
    sum=0
    for i in range(0,n):
        sum=sum+i
    print("sum of", n ,"no's are:",sum)
def mul(n):
    mul=1
    for i in range(1,n+1):
        mul=mul*i
    print("Multiplication of", n ,"no's are:",mul) 
    
n=int(input("Enter the number:"))
print("1: To Get Sum")
print("2: To Get Product")
Choice=int(input("Choose either 1 or 2:"))
if Choice==1:
    sum(n)
elif Choice==2:
    mul(n)
else: 
    print("INVALID INPUT")
 
```
